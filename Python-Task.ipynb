```python
from bs4 import BeautifulSoup
import requests 
 
# Function to extract Product Title
def get_title(soup):
     
    try:
        # Outer Tag Object
        title = soup.find("span", attrs={"id":'productTitle'}).string.strip()
        
 
    except AttributeError:
        title = ""   
 
    return title
 
# Function to extract Product Price
def get_price(soup):
 
    try:
        price = soup.find("span", attrs={'class':'a-price-whole'}).text.strip()
 
    except AttributeError:
        price = ""  
 
    return price

 
# Function to extract the product of details
def get_details(soup):
 
    try:
        details = soup.find('ul',attrs={'class':'a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list'}).text.strip()
 
    except AttributeError:
        details = ""  
 
    return details

 
if __name__ == '__main__':
 
    # Headers for request
    HEADERS = ({'User-Agent':
                'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
                'Accept-Language': 'en-US, en;q=0.5'})
 
    # The webpage URL
    URL = "https://www.amazon.in/de/dp/000004458X"
 
    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)
 
    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "lxml")
 
    # Function calls to display all necessary product information
    print("Product Title =", get_title(soup))
    print("Product Price =", get_price(soup))
    print("Product details =", get_details(soup).strip())
    
    
# Function to extract the product image urls
def getdata(url): 
    r = requests.get(url) 
    return r.text 
    
htmldata = getdata("https://www.amazon.in/de/dp/000004458X") 
soup = bs(htmldata, 'html.parser') 
for item in soup.find_all('img'):
    print(item['src'])
```
